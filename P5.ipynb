```python
X = np.linspace(0, 10, 100)
y = 2 * X + 1 + np.random.randn(100)
m, b = 0, 0
L = 0.001
epochs = 1000
n = float(len(X))
for i in range(epochs):
    y_pred = m * X + b
    D_m = (-2/n) * sum(X * (y - y_pred))
    D_b = (-2/n) * sum(y - y_pred)
    m = m - L * D_m
    b = b - L * D_b
y_fit = m * X + b
print("5. Gradient Descent Results")
print("Slope:", round(m,3), "Intercept:", round(b,3))
plt.scatter(X, y)
plt.plot(X, y_fit, color='red')
plt.title("Gradient Descent Line Fit")
plt.show()
```
